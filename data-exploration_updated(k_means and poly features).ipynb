## Feature engineering

### 1 .determine the features and target variables

```python
X = data.drop(["loneliness","sporting"],axis = 1)
y_loneliness = data["loneliness"]
y_sporting = data["sporting"]
```

### 2. Using KNN to handle missing Value. 

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors = 30)
X = imputer.fit_transform(X)
```

### 3  Scale the data 

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler, MinMaxScaler
from typing import Union
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer

standard = StandardScaler()
minmax = MinMaxScaler()
```

#### 3.1 Data that scale by standard scaler

```python
from sklearn.model_selection import train_test_split
```

```python
X_sd = standard.fit_transform(X)

X_lone_train_sd, X_lone_test_sd, y_lone_train, y_lone_test = train_test_split(
    X_sd, y_loneliness, random_state=100)

X_sport_train_sd, X_sport_test_sd, y_sport_train, y_sport_test = train_test_split(
    X_sd, y_sporting, random_state=100)
```

#### 3.2 Data that scales by max_min scaler

```python
X_mm = minmax.fit_transform(X)

X_lone_train_mm, X_lone_test_mm, y_lone_train, y_lone_test = train_test_split(
    X_mm, y_loneliness, random_state=100)

X_sport_train_mm, X_sport_test_mm, y_sport_train, y_sport_test = train_test_split(
    X_mm, y_sporting, random_state=100)
```

### 4. Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures


#define the polynomial features of X_lone set
poly = PolynomialFeatures(degree = 2).fit(X_mm)
X_poly = poly.transform(X_mm)


X_lone_train_poly = poly.transform(X_lone_train_mm)
X_lone_test_poly  = poly.transform(X_lone_test_mm)


#define the polynomial features of X_sport set

X_sport_train_poly = poly.transform(X_sport_train_mm)
X_sport_test_poly  = poly.transform(X_sport_test_mm)
```

##### 2.2.1 Find the best K value

```python
# Find the best K value

from sklearn.cluster import KMeans
ks = range(1,10)
inertias = [] #distance

for k in ks:
    
    model = KMeans(n_clusters = k)
    model.fit(X_poly)
    inertias.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,6))
plt.style.use("bmh")
plt.plot(ks, inertias)
plt.xlabel("K value")
plt.ylabel("Distance")
```

We found that the rate of decline is the fastest when k=2, so k is 2 is a better value

Silhouette Method：(verify)

```python
from sklearn.metrics import silhouette_samples, silhouette_score

for k in range(2,5):
    fig,ax = plt.subplots(1,1)
    fig.set_size_inches(18,7)
    
    km = KMeans(n_clusters = k, random_state = 0)
    labels = km.fit_predict(X_poly)
    
    silhouette_values = silhouette_samples(X_poly, labels)
    
    
    y_lower, y_upper = 0,0
    
    
    for index, cluster in enumerate(np.unique(labels)):
        
        cluster_silhouette_values = silhouette_values[labels == cluster]
        cluster_silhouette_values.sort()
        
        y_upper += len(cluster_silhouette_values)
        
        ax.barh(range(y_lower, y_upper), cluster_silhouette_values,height = 1)
        ax.text(-0.05, (y_lower+y_upper)/2,index)
        
        y_lower += len(cluster_silhouette_values)
        
        avg_score = np.mean(silhouette_values)
        ax.axvline(avg_score, linestyle ="--", linewidth = 2, color = 'green')
        ax.set_yticks([])
        
        ax.set_xlabel("Silhouette Coefficient Values")
        ax.set_ylabel("Cluster Labels")
        ax.set_title("Silhouette plot for the various clusters")
```

ps:the avg score is highest when k =2 clearly

##### 2.2.2 Create a table to show results

```python
def k_means(label_1: List, label_2: List, 
            data_frame:np.array = X_poly, 
            n_clusters:int = 2)-> None:
    kmeans_model = KMeans(n_clusters = n_clusters, n_init = 30, random_state = 0)
    kmeans_model.fit(data_frame)
    k_labels= kmeans_model.labels_

    df = pd.DataFrame({"cluster_label": k_labels, "Feeling of Loneliness": label_1, "Level of Sporting": label_2})
    ct_1 = pd.crosstab(df["cluster_label"], df["Feeling of Loneliness"])
    ct_2 = pd.crosstab(df["cluster_label"], df["Level of Sporting"])
    
    display(ct_1)
    display(ct_2)
```

#### Define a function that describes the level of activity

```python
from typing import List

def change_label(variables:List[float])->List[str]:
    l = []

    for i in range(len(variables)):
        if variables[i] < 25:
            l.append("Seldom")

        elif variables[i] <50:
            l.append("Sometimes")

        elif variables[i] <75:
            l.append("Often")

        else:
            l.append("Frequent")

    return l
```

```python
f_lone = change_label(y_loneliness)
l_sport =  change_label(y_sporting)
```

```python
k_means(label_1 = f_lone,label_2 = l_sport)
```
