## define a function that could encode and scale data.

### 2.determine the features and target variables

```python
X = data.drop(["loneliness","sporting"],axis = 1)
y_loneliness = data["loneliness"]
y_sporting = data["sporting"]
```

### 2.1 Using KNN to handle missing Value. 

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors = 30)
X = imputer.fit_transform(X)
```

### 3  scale the data 

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler, MinMaxScaler
from typing import Union
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer

standard = StandardScaler()
minmax = MinMaxScaler()
```

### 4. Data that scale by standard scaler

```python
from sklearn.model_selection import train_test_split
```

```python
X_sd = standard.fit_transform(X)

X_lone_train_sd, X_lone_test_sd, y_lone_train, y_lone_test = train_test_split(
    X_sd, y_loneliness, random_state=100)
```

## Unsupervised method

### Dependence between variables (Determine the variance-covariance matrix)

### find PCA

```python
from sklearn.decomposition import PCA
pca_X = PCA(n_components=5)# reduce dem to 5
principalComponents_X = pca_X.fit_transform(X_sd)

principal_X_Df = pd.DataFrame(data = principalComponents_X
             , columns = ['principal component 1', 'principal component 2','principal component 3',
                          'principal component 4','principal component 5'])
```

```python
principal_X_Df.head()
```

#### for loneliness

```python
plt.scatter(principalComponents_X[:, 0], principalComponents_X[:, 1],
            c=y_loneliness, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Accent', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar()
```

#### for soprting

```python
plt.scatter(principalComponents_X[:, 0], principalComponents_X[:, 1],
            c=y_sporting, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Accent', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar()
```

```python
import seaborn as sns
df_cm = pd.DataFrame(np.abs(pca_X.components_),columns = data.columns[:-2])
plt.figure(figsize = (150,10))
ax = sns.heatmap(df_cm, annot=True, cmap="BuPu")

ax.yaxis.set_tick_params(labelsize=1)
ax.xaxis.set_tick_params(labelsize=15)
plt.title('PCA', fontsize='xx-large')
# Set y-axis label
plt.savefig('factorAnalysis.png', dpi=200)
```
